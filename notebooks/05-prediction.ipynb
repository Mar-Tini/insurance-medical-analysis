```python
import init_notebooks 
from data_loader import load_data

X_train, X_test, y_train, y_test = load_data()
```

```python
import joblib
loaded_model = joblib.load("../models/best_rf_model.pkl")
```

Charge le modèle sauvegardé qui contient le préprocesseur + le XGBRegressor entraîné avec les meilleurs paramètres.

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = loaded_model.predict(X_test)
print("Prediction:", loaded_model.score(X_test, y_test))
print("Test MSE:", mean_squared_error(y_test, y_pred))
print("Test R²:", r2_score(y_test, y_pred))
```

le modèle explique environ 88 % de la variance des charges.

R² élevé et MSE relativement faible, modèle robuste et fiable pour ce dataset.

Étape finale avant sauvegarde ou utilisation en production

```python
import joblib
import matplotlib.pyplot as plt

y_pred = loaded_model.predict(X_test)

plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         'r--', lw=2) 
plt.xlabel("Valeurs réelles")
plt.ylabel("Prédictions")
plt.title("Prédictions vs Valeurs réelles")
plt.show()
```
